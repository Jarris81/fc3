```python
import sys
import pandas as pd
# sys.executable # should be thesis env...
```

```python
# merge tower experiments
df_tower_simple = pd.read_csv("experiments_run_1_tower_simple.csv", index_col=False)
df_tower_rlds_clone = pd.read_csv("experiments_run_1_tower_rlds_clone.csv", index_col=False)
df_tower_rlgs = pd.read_csv("experiments_run_1_tower_rlgs.csv", index_col=False)
```

```python
df_tower = pd.concat((df_tower_simple, df_tower_rlds_clone, df_tower_rlgs), axis=0).reset_index(drop=True)
df_tower
```

```python
df_stick = pd.read_csv("experiments_run_1_stick.csv", index_col=False)
df_hand_over = pd.read_csv("experiments_run_1_hand_over.csv", index_col=False)
```

```python
df_all = pd.concat((df_tower, df_stick, df_hand_over), axis=0).reset_index(drop=True)
df_all.to_csv("experiment_run_all_clean.csv", index=False)
```
